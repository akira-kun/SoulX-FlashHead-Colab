### Baixar pesos do X-NeMo diretamente do Google Drive usando `gdown`

Esta seГ§ГЈo demonstra como baixar diretamente os arquivos da pasta do Google Drive fornecida. Este mГ©todo Г© uma alternativa ao upload manual ou Г  montagem do Google Drive, especialmente Гєtil se vocГЄ nГЈo tiver os arquivos no seu Drive.

```python
# Instalar gdown, se ainda nГЈo estiver instalado
!pip install -q gdown

# ID da pasta do Google Drive fornecida
FOLDER_ID = '1RdjBYYbstO7SOchDg7oimoAwu03g_-mI'

# Crie a pasta de destino para os pesos do X-NeMo
target_dir = '/content/x-nemo-inference/pretrained_weights/x-nemo-model'
!mkdir -p {target_dir}

print(f"Baixando arquivos da pasta do Google Drive ({FOLDER_ID}) para {target_dir}...")

# Mude para a pasta de destino antes de baixar
%cd {target_dir}

# Baixar os arquivos da pasta do Google Drive usando gdown
# '--folder' indica que estamos baixando uma pasta. Como jГЎ estamos no diretГіrio certo, nГЈo precisamos de -P.
!gdown --folder --id {FOLDER_ID}

# Volte para o diretГіrio base do projeto
%cd /content/x-nemo-inference

print("Download dos pesos do X-NeMo concluГ­do com sucesso!")
```
